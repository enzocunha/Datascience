# Redes neurais

## Introdução

Redes neurais são uma categoria de algoritmos de machine learning inspirados pelo cérebro humano, tentando imitar como os neurônios se comunicam.

A ideia por tras do funcionamento desses modelos é que há diversas camadas com diferentes nós que podem receber e transmitir informação associados entre si, assim, quando um nó recebe um dado, se o peso passar de uma barreira definida, ele transmite sua saída para a próxima camada.

## Vantagens e desvantagens

### Vantagens
* Habilidade de aprender modelos não lineares complexos e aprender em tempo real
* Falha em um nó não faz com que a rede não gere uma saída
### Desvantagens
* Caixa preta: Díficil entender porque ocorreu uma decisão
* Necessita de grandes volumes de dados
* Redes e datasets e complexos tendem a necessitar muito tempo de treinamento

## Base matemática e algoritmo

Há alguns tipos de redes neurais, mas pensando em uma rede MLP (ou Perceptron Multicamadas), cada nó individualmente pode ser comparado a um modelo de regressão linear, possuindo uma entrada, peso, viés e saída.

A função para um nó é algo como:

<img src='https://1.cms.s81c.com/sites/default/files/2021-01-06/ICLH_Diagram_Batch_01_05B-Equations-WHITEBG.png' width=400>

sendo:
* w é o peso atribuido quando a camada de entrada é determinada
* x é a entrada

Então caso a função ultrapasse a barreira definida, o nó passa essa saída para a próxima camada.

### Etapas

1. Dados são passados para a camada de entrada, que atribui pesos (inicialmente aleatórios) e os passa para a próxima camada
2. Os nós dessas camadas são associados a um valor numérico (viés) que é somado ao valor recebido da camada anterior e passado para uma função de ativação
3. O resultado da função de ativação determina se aquele neurônio em particular é ativado, passando os dados para a próxima camada com seu respectivo peso
4. Após passar pelas camadas ocultas, chegando à camada de saída, o resultado é definido pelo maior valor entre os nós, representando a maior probabilidade de um resultado correto

### Tipos de redes neurais

Há diferentes tipos de redes neurais, e cada tipo pode ser mais eficiente para um tipo de problema. Alguns tipos de redes neurais são apresentados a seguir:

* Perceptron: Tipo mais antigo de rede neural composta por apenas um neurônio.

* Perceptron Multicamadas ou Feedforward: Foco desse trabalho. Essa rede é composta de uma camada de entrada, uma ou mais camadas ocultas e uma camada de saída, mas diferente do que o nome sugere, os neurônios não são perceptrons.

* Rede neural convolucional: Estrutura similar à anterior, porém utilizando princípios de algebra linear e adaptada para problemas envolvendo matrizes, como processamento de imagens.

* Rede neural recorrente: Possuem loops de feedback e são normalmente utilizadas para problemas temporais.

### Redes neurais profundas

Esse tipo de rede neural que foi teorizada na década de 80 do século passado, mas tem sido foco de muita atenção e estudos nessa última década e há 2 grandes motivos para isso:

* Necessita de enormes volumes de dados. O desenvolvimento de carros que dirigem sozinhos requerem milhões de imagens e milhares de horas de vídeo.
* Requer grande poder computacional que pode ser atingido atualmente por GPUs e clusters.

As redes neurais profundas são essencialmente uma rede neural com 3 ou mais camadas. Essas camadas extras em relação a uma rede neural qualquer permitem às redes neurais profundas refinar e otimizar seus resultados.

<img src='https://1.cms.s81c.com/sites/default/files/2021-04-22/Russian%20Nesting%20Dolls.png' width=500>

Essas redes tem a capacidade de fazer pré processamento em dados não estruturados, como textos e imagens, e extrair as features mais importantes para a classificação, algo que normalmente deve ser feito manualmente por pessoas para depois alimentar o algoritmo de machine learning.

## Exemplo


Nesse exemplo é utilizado o dataset "Optical Recognition of Handwritten Digits" fornecido pelo scikit learn.

Esse dataset possui 1797 exemplos de imagens 8x8 (64 dimensões) e o objetivo é identificar o digito entre 0 e 16.


```python
from sklearn.datasets import load_digits
import pandas as pd

data = load_digits()
df = pd.DataFrame(data=data.data, columns=data.feature_names)
df.head()
```

### Modelo

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split

x_treino, x_teste, y_treino, y_teste = train_test_split(data.data, data.target, test_size=0.3, random_state=0)

clf = MLPClassifier()
clf.fit(x_treino, y_treino)
predicao = clf.predict(x_teste)
```

### Resultados

```python
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report

print(classification_report(y_teste, predicao))

matriz_confusao = ConfusionMatrixDisplay.from_estimator(
    clf,
    x_teste,
    y_teste,
    cmap=plt.cm.Blues,
    normalize=None,
)
matriz_confusao.ax_.set_title('Matriz de confusão')

plt.show()
```
